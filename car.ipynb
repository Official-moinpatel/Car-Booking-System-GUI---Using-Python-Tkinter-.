```python
import tkinter as tk
from tkinter import messagebox, ttk

def confirm_booking():
    name = name_entry.get()
    phone = phone_entry.get()
    car = car_combobox.get()
    pickup = pickup_entry.get()
    return_date = return_entry.get()

    if not (name and phone and car and pickup and return_date):
        messagebox.showwarning("Input Error", "All fields are required!")
        return

    summary = f"""
    Booking Confirmed!

    Name: {name}
    Phone: {phone}
    Car Selected: {car}
    Pickup Date: {pickup}
    Return Date: {return_date}
    """
    messagebox.showinfo("Booking Summary", summary)

    # Optional: Save booking to file
    with open("car_bookings.txt", "a") as file:
        file.write(summary + "\n" + "-"*40 + "\n")

    # Clear form
    name_entry.delete(0, tk.END)
    phone_entry.delete(0, tk.END)
    pickup_entry.delete(0, tk.END)
    return_entry.delete(0, tk.END)
    car_combobox.set("")

# Main window setup
root = tk.Tk()
root.title("Car Booking System")
root.geometry("400x400")
root.config(bg="#f0f0f0")

# Title label
tk.Label(root, text="Car Booking Form", font=("Arial", 16, "bold"), bg="#f0f0f0").pack(pady=10)

# Name
tk.Label(root, text="Customer Name:", bg="#f0f0f0").pack(anchor="w", padx=20)
name_entry = tk.Entry(root, width=30)
name_entry.pack(padx=20)

# Phone
tk.Label(root, text="Phone Number:", bg="#f0f0f0").pack(anchor="w", padx=20, pady=(10, 0))
phone_entry = tk.Entry(root, width=30)
phone_entry.pack(padx=20)

# Car selection
tk.Label(root, text="Select Car Model:", bg="#f0f0f0").pack(anchor="w", padx=20, pady=(10, 0))
car_combobox = ttk.Combobox(root, values=["Honda City", "Toyota Innova", "Hyundai i20", "Maruti Swift"], state="readonly", width=27)
car_combobox.pack(padx=20)

# Pickup date
tk.Label(root, text="Pickup Date (DD-MM-YYYY):", bg="#f0f0f0").pack(anchor="w", padx=20, pady=(10, 0))
pickup_entry = tk.Entry(root, width=30)
pickup_entry.pack(padx=20)

# Return date
tk.Label(root, text="Return Date (DD-MM-YYYY):", bg="#f0f0f0").pack(anchor="w", padx=20, pady=(10, 0))
return_entry = tk.Entry(root, width=30)
return_entry.pack(padx=20)

# Submit button
tk.Button(root, text="Confirm Booking", command=confirm_booking, bg="#007acc", fg="white", font=("Arial", 12, "bold")).pack(pady=20)

# Run the app
root.mainloop()
```
